```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
data = datasets.load_iris()
X = data.data
y = data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

### Gini Index (default)

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Accuracy is : ",accuracy_score(y_train,classifier.predict(X_train)))
print("Accuracy is : ",accuracy_score(y_pred,y_test))
```

#### Entropy

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Accuracy (entropy) is : ",accuracy_score(y_train,classifier.predict(X_train)))
print("Accuracy (entropy) is : ",accuracy_score(y_pred,y_test))
```

###Visualization of Decision Tree

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data = StringIO()
```

```python
export_graphviz(classifier,out_file=dot_data,class_names=data.target_names,feature_names=data.feature_names,special_characters=True,filled=True,rounded=True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
